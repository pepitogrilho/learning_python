```python
def divide(a, b):
  if b==0:
    raise ValueError("Division by zero")
  return a / b

try:
  result = divide(10, 0)
except ValueError as e:
  print("Failed:", e)
```

```python
def shout(word):
    """
    Print a word with an exclamation mark

    Parameters
    ----------
    word : string
        Word to be shouted.

    Returns
    -------
    The shouted word.

    """
    return word + "!!!!!!"

shout("Basta")
print(shout("Basta"))
```
